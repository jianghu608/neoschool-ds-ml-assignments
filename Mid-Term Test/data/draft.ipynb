```python
pi = 0
term_sum = 0
term_num = 0
rela_err = 1

def approach_pi():
    global term_sum, term_num, pi, rela_err

    term_n = ((-1) ** term_num) / (2 * term_num + 1)
    term_sum += term_n
    pi = term_sum * 4
    rela_err = abs((pi - np.pi) / np.pi)
    term_num += 1

def main():
    while (rela_err >= 1e-6):
        approach_pi()

    print(pi, term_num, rela_err)

main()
```

```python
pi = 0
term_sum = 0
term_num = 0
rela_err = 1

for i in range(20):
    approach_pi()
    print(pi, term_num, rela_err)
```
