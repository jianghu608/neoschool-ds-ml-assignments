## More on accessing files using Python

### Path - identify the location of a file

In Python, we normally use a `str` to encode a path.

#### Absolute path - the absolute location of a file

- Windows: `C:\Users\xuheng\Documents\data.csv`
    - Begins with the disk label, e.g. `C:\`
    - Uses a back-slash(`\`, the key above your `Enter` key) to separate directory structures
- Unix-like: `/home/xuheng/Documents/data.csv`
    - Begins with a slash(`/`, the same key as the `?` key)
    - Uses a slash(`/`) to separate  directory structures

Python handles `/` and `\` automatically for us, so we can use them interchangeably.

We will use `/` for the rest of the class.

## IMDB dataset

```python
df.head()
```

**We need to clean up some of the data**

```python
df_clean = df.copy()
df_clean['Runtime'] = df_clean['Runtime'].str.replace(' min', '').astype(int)
df_clean['Gross'] = pd.to_numeric(df_clean['Gross'].str.replace(',', ''), errors='coerce')
df_clean['Released_Year'] = pd.to_numeric(df_clean['Released_Year'], errors='coerce')
df_clean['Primary_Genre'] = df_clean['Genre'].str.split(',').str[0]
```
